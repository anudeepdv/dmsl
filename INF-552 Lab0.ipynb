#### The iterrows() Method

Create a Python dictionary object whose keys are the headers of the dataframe created in the read_csv() exercise and values are Python list objects that contain data corresponding to the headers. (Here, use the iterrows method to iterate each row of the dataframe and copy it to a dictionary. However, there is a easier way. Learn how the to_dict() method works by yourself later)

````python
data_dict = {}

for i, row in data.iterrows():
    for c in data.columns:
        if c not in data_dict:
            data_dict[c] = []
        data_dict[c].append(row[c])
        
data_dict
````

(d) Create multiple subplots, play around with the figure size, figure title, and its font style and font size **(One curve is plotted in one subplot in this question)**.


copy the code above to the following cell and add code for plotting curve $\left(x, y1\right)$, $\left(x, y2\right)$ and $\left(x, y3\right)$.

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, num=20)
y1 = np.array([j for j in x])
y2 = np.array([j ** 2 for j in x])
y3 = np.array([j ** 3 for j in x])

# Create a figure with multiple subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # 1 row, 3 columns of subplots

# Plot the first curve (x, y1) in the first subplot
axes[0].plot(x, y1, label='y = x', color='blue')
axes[0].set_title('Subplot 1', fontsize=14, fontweight='bold')
axes[0].set_xlabel('x', fontsize=12)
axes[0].set_ylabel('y', fontsize=12)
axes[0].legend()

# Plot the second curve (x, y2) in the second subplot
axes[1].plot(x, y2, label='y = x^2', color='green')
axes[1].set_title('Subplot 2', fontsize=14, fontweight='bold')
axes[1].set_xlabel('x', fontsize=12)
axes[1].set_ylabel('y', fontsize=12)
axes[1].legend()

# Plot the third curve (x, y3) in the third subplot
axes[2].plot(x, y3, label='y = x^3', color='red')
axes[2].set_title('Subplot 3', fontsize=14, fontweight='bold')
axes[2].set_xlabel('x', fontsize=12)
axes[2].set_ylabel('y', fontsize=12)
axes[2].legend()

# Set a title for the entire figure
fig.suptitle('Multiple Subplots with Different Curves', fontsize=16, fontweight='bold')

# Adjust spacing between subplots
plt.tight_layout()

# Show the figure
plt.show()
````

(e) Change the limits on x and y axes, **use logarithmic axes to plot**.

```python

```

copy the code above to the following cell and add code for plotting the parabola.

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, num=20)
y = np.array([j ** 2 for j in x])

# Create a figure with a logarithmic x-axis and y-axis
plt.figure(figsize=(8, 6))
plt.semilogx(x, y, label='y = x^2', color='blue')  # Logarithmic x-axis
plt.semilogy(x, y, label='y = x^2', color='green')  # Logarithmic y-axis

# Set custom limits for the x and y axes
plt.xlim(0.01, 10)  # Custom x-axis limits (log scale)
plt.ylim(0.01, 100)  # Custom y-axis limits (log scale)

# Add labels and legend
plt.xlabel('log(x)', fontsize=12)
plt.ylabel('log(y)', fontsize=12)
plt.legend()

# Set a title for the figure
plt.title('Logarithmic Axes with Custom Limits', fontsize=14, fontweight='bold')

# Show the plot
plt.grid(True)  # Add gridlines
plt.show()
````
